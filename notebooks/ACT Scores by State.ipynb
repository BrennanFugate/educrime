```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
scores = pd.read_csv('../data/ACT_Scores_by_State.csv').dropna()
scores
```

```python
scores.columns.drop('RegionDROP')
```

```python
scores = scores.drop(['RegionDROP'], axis=1)
```

```python
scores
```

```python
scores['State'] = scores['State'].astype('string')
```

```python
scores['Region'] = scores['Region'].astype('string')
```

```python
scores.info()
```

```python
state_regions = scores[['Region', 'State']]
state_regions
```

```python
state_regions = state_regions.groupby('Region').count()
```

```python
state_regions = state_regions.reset_index()
state_regions
```

```python
state_regions.info()
```

```python
sns.histplot(data = state_regions)
```
